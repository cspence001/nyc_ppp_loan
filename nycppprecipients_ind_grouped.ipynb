```python
import pandas as pd
import numpy as np
```

```python
#file path
csv_path = "data/nycppprecipients_split.csv"
```

```python
#import csv to df
ppp_df = pd.read_csv(csv_path, encoding="ISO-8859-1")
ppp_df.head()
```

```python
conditions = [
    (ppp_df['NA'] == 11),
    (ppp_df['NA'] == 21),
    (ppp_df['NA'] == 22),
    (ppp_df['NA'] == 23),
    (ppp_df['NA'] == 31),
    (ppp_df['NA'] == 32),
    (ppp_df['NA'] == 33),
    (ppp_df['NA'] == 42),
    (ppp_df['NA'] == 44),
    (ppp_df['NA'] == 45),
    (ppp_df['NA'] == 48),
    (ppp_df['NA'] == 49),
    (ppp_df['NA'] == 51),
    (ppp_df['NA'] == 52),
    (ppp_df['NA'] == 53),
    (ppp_df['NA'] == 54),
    (ppp_df['NA'] == 55),
    (ppp_df['NA'] == 56),
    (ppp_df['NA'] == 61),
    (ppp_df['NA'] == 62),
    (ppp_df['NA'] == 71),
    (ppp_df['NA'] == 72),
    (ppp_df['NA'] == 81),
    (ppp_df['NA'] == 92),
]

values = ['Agriculture, Forestry, Fishing and Hunting', 'Mining', 
          'Utilities', 'Construction', 'Manufacturing', 'Manufacturing', 
         'Manufacturing', 'Wholesale Trade', 'Retail Trade', 'Retail Trade',
         'Transportation and Warehousing', 'Transportation and Warehousing',
         'Information', 'Finance and Insurance', 'Real Estate Rental and Leasing',
         'Professional, Scientific, and Technical Services', 'Management of Companies and Enterprises',
         'Administrative and Support and Waste Management and Remediation Services',
         'Educational Services', 'Health Care and Social Assistance', 'Arts, Entertainement, and Recreation',
         'Accomodation and Food Services', 'Other Services (except Public Administration)',
         'Public Administration']

ppp_df['Industry'] = np.select(conditions, values)

ppp_df.head()
```

```python
#columns re-ordered and cleaned
ppp_df = ppp_df [["LoanRange", "BusinessName", "Address", "City", "State",
                 "Zip", "NAICSCode", "NA", "Industry", "Description", "BusinessType", "JobsRetained",
                 "Lender", "CD"]]
ppp_df.head()
```

```python
ppp_df.to_csv("data/output/nycppprecipients_ind.csv", index=True)
```

```python
#range count all industry
ind_range = ppp_df["LoanRange"].value_counts()
ind_range
```

```python
#range count per industry
ind_group = ppp_df.groupby(["Industry"])
rng_count = ind_group["LoanRange"].value_counts()
print(rng_count)
```

```python
#for sum of jobs, count of loans based on loan range per industry 
jobsByRange_df = ppp_df.groupby(["Industry", "LoanRange"]).agg(CountLoanRange=('LoanRange', 'count'),SumJobsRetained=('JobsRetained', 'sum'))
jobsByRange_df
```

```python
#loan counts per industry
ind_count = ppp_df["Industry"].value_counts()
ind_count
```

```python
#job counts per industry
ppp_df_ind = ppp_df.groupby(["Industry"])
ppp_df_ind_comp = ppp_df_ind["JobsRetained"].sum() 
ppp_df_ind_comp
```
