```python
#imports
import requests
import pandas as pd
import numpy as np
from tqdm.notebook import tqdm 
```

```python
#import csv to df
ppp_ind_df = pd.read_csv(csv_path, encoding="ISO-8859-1")
ppp_ind_df.head()
```

```python
ppp_ind_df['full_add'] = ppp_ind_df.Address + ", " + ppp_ind_df.City + ", " + ppp_ind_df.State
ppp_ind_df
```

```python
params = {
    "key":API_KEY
}
base_url = 'https://maps.googleapis.com/maps/api/geocode/json?'
addresses = ppp_ind_df["full_add"]
lat = []
lng = []
notfound = []
for address in tqdm(addresses):
    params['address'] = f"{address}"
    response = requests.get(base_url, params=params).json()
    results = response['results']
    try:
        print(f"{address} is {results[0]['geometry']['location']}")
        #geometry = results[0]['geometry']
        lat.append(results[0]['geometry']['location']['lat'])
        lng.append(results[0]['geometry']['location']['lng'])
        #print(lat, lng)
    except (KeyError, IndexError):
        print(f"{address}Not Found")
        notfound.append(address)
        lat.append(None)
        lng.append(None)
```
